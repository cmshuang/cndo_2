```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from mpl_toolkits.mplot3d import Axes3D
```

```python
def parse_atom_positions(filename):
    """
    File are of format:
    num_atoms
    A1 x1 y1 z1
    A2 x2 y2 z2
    ...
    Ak xk yk zk
    """
    with open(filename) as f:
        num_atoms = float(f.readline().strip().split()[0])
        
        positions_L = []
        for line in f:
            xyz = np.array([float(x_i) for x_i in line.strip().split()[1:]])
            positions_L.append(xyz)
            
        positions = np.stack(positions_L)
    assert(num_atoms == positions.shape[0])
    return np.transpose(positions)
    
```

## Results of optimization

```python
# Parse files containing atom information

# First test case, 2 atoms
starting_pos_1 = parse_atom_positions("./input/H2.txt")
final_pos_1 = parse_atom_positions("./output.txt")

# We choose a starting point where the distance between the two atoms is close to the expected optimal bond length of 5.29A (sigma=5.29A)
print(starting_pos_1)
# The distance between these two atoms is 5.35 A
```

```python
%matplotlib inline
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# draw a sphere for each data point

ax.scatter(starting_pos_1[0],starting_pos_1[1],starting_pos_1[2], color='gold', alpha=0.9, s=100)
    
plt.title("Initial Particle Positions, Case 1")
plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# draw a sphere for each data point

ax.scatter(final_pos_1[0],final_pos_1[1],final_pos_1[2], color='gold', alpha=0.9, s=100)
    
plt.title("Final Particle Positions, Case 1")
ax.view_init(30, 120)
plt.show()
print(final_pos_1)
```
